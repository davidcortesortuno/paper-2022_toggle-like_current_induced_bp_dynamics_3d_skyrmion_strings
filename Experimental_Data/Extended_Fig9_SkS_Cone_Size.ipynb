```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as c
import scipy.fftpack as fft
import scipy.ndimage as ndi
import scipy as sp
from PIL import Image
import csv
def px_remove(array, x, y):
    temp = array
    temp[y,x]=(temp[y+1,x]+temp[y-1,x]+temp[y,x+1]+temp[y,x-1])/4
    return temp

def px_remove_edge(array, x, y):
    temp = array
    temp[y,x]=(temp[y+1,x]+temp[y-1,x]+temp[y,x-1])/3
    return temp

plt.rcParams.update({'font.size':13})
plt.rcParams['font.sans-serif'] = "Helvetica"
```

```python
### Energy Spectra ###
%matplotlib notebook

fs = [65,70,75,80,85,90,95]

Cone_list = []
SkT_list = []

Cone_px = []
SkT_px = []

for i in np.arange(0,len(fs),1):
    Cone=np.loadtxt('STXM_Data/SkS_Together_Field_sweep_up/Line_Scans/Cone_'+str(fs[i])+'mT.csv',dtype='float',skiprows=1,delimiter=',')
    SkT=np.loadtxt('STXM_Data/SkS_Together_Field_sweep_up/Line_Scans/SkT_'+str(fs[i])+'mT.csv',dtype='float',skiprows=1,delimiter=',')
    
    Cone_list.append(Cone[:,1])
    SkT_list.append(SkT[:,1])
    
    Cone_px.append(Cone[:,0]*15)
    SkT_px.append(SkT[:,0]*15)
```
