```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as c
import scipy.fftpack as fft
import scipy.ndimage as ndi
import scipy as sp
from PIL import Image

def px_remove(array, x, y):
    temp = array
    temp[y,x]=(temp[y+1,x]+temp[y-1,x]+temp[y,x+1]+temp[y,x-1])/4
    return temp

def px_remove_edge(array, x, y):
    temp = array
    temp[y,x]=(temp[y+1,x]+temp[y-1,x]+temp[y,x-1])/3
    return temp
```

# Load Data

```python
%matplotlib notebook

fields = ['n250','n200','n150','n100','n75','n50','n25','p0','p25','p50','p75','p100','p150','p200','p250',]

OOP_list = []

for i in np.arange(0,len(fields),1):
    pos_im = Image.open("STXM_Data/OOP_Field_Sweep/TIF_Data/OOP_"+fields[i]+"mT_pos.tif")
    neg_im = Image.open("STXM_Data/OOP_Field_Sweep/TIF_Data/OOP_"+fields[i]+"mT_neg.tif")
    pos =np.array(pos_im)
    neg =np.array(neg_im)

    pos = np.rot90(pos)
    neg = np.rot90(neg)

    final = pos/neg
    final = np.fliplr(final)
    OOP_list.append(final)
```

```python
OOP_final_list = []
# -250 mT #
temp1 = px_remove(OOP_list[0], 170, 19)
temp2 = px_remove(temp1, 170, 20)
OOP_final_list.append(temp2)

# -200 mT #
OOP_final_list.append(OOP_list[1])

# -150 mT #
temp1 = px_remove(OOP_list[2], 26, 21)
OOP_final_list.append(temp1)

# -100 mT #
OOP_final_list.append(OOP_list[3])

# -75 mT #
OOP_final_list.append(OOP_list[4])

# -50 mT #
temp1 = px_remove(OOP_list[5], 100, 7)
OOP_final_list.append(temp1)

# -25 mT #
OOP_final_list.append(OOP_list[6])

# 0 mT #
OOP_final_list.append(OOP_list[7])

# 25 mT #
temp1 = px_remove(OOP_list[8], 139, 22)
OOP_final_list.append(temp1)

# 50 mT #
OOP_final_list.append(OOP_list[9])

# 75 mT #
temp1 = px_remove(OOP_list[10], 102, 29)
OOP_final_list.append(temp1)

# 100 mT #
temp1 = px_remove(OOP_list[11], 66, 8)
temp2 = px_remove(temp1, 99, 9)
OOP_final_list.append(temp2)

# 150 mT #
OOP_final_list.append(OOP_list[12])

# 200 mT #
OOP_final_list.append(OOP_list[13])

# 250 mT #
OOP_final_list.append(OOP_list[14])
```
