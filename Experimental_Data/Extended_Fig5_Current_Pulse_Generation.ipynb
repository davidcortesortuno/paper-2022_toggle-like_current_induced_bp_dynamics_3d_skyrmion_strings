```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as c
import scipy.fftpack as fft
import scipy.ndimage as ndi
import scipy as sp
from PIL import Image

def px_remove(array, x, y):
    temp = array
    temp[y,x]=(temp[y+1,x]+temp[y-1,x]+temp[y,x+1]+temp[y,x-1])/4
    return temp

def px_remove_edge(array, x, y):
    temp = array
    temp[y,x]=(temp[y+1,x]+temp[y-1,x]+temp[y,x-1])/3
    return temp
```

# Load Data

```python
%matplotlib notebook

fields = ['0','10','20','30','40','50','60','70','80']

Gen_list = []

for i in np.arange(0,len(fields),1):
    pos_im = Image.open("STXM_Data/Pulse_Different_IP_Fields/TIF_Data/Pulse_"+fields[i]+"mT_pos.tif")
    neg_im = Image.open("STXM_Data/Pulse_Different_IP_Fields/TIF_Data/Pulse_"+fields[i]+"mT_neg.tif")
    pos =np.array(pos_im)
    neg =np.array(neg_im)

    pos = np.rot90(pos)
    neg = np.rot90(neg)

    final = pos/neg
    final = np.fliplr(final)
    Gen_list.append(final)
```

```python
Gen_final_list = []
temp1 = px_remove(Gen_list[0], 148, 24)
#temp2 = px_remove(temp1, 68, 23)
Gen_final_list.append(temp1)

temp1 = px_remove(Gen_list[1], 177, 19)
temp2 = px_remove(temp1, 134, 15)
Gen_final_list.append(temp2)

temp1 = px_remove(Gen_list[2], 156, 20)
temp2 = px_remove(temp1, 39, 23)
Gen_final_list.append(temp2)

temp1 = px_remove(Gen_list[3], 49, 24)
#temp2 = px_remove(temp1, 39, 23)
Gen_final_list.append(temp1)

temp1 = px_remove(Gen_list[4], 93, 12)
#temp2 = px_remove(temp1, 39, 23)
Gen_final_list.append(temp1)

temp1 = px_remove(Gen_list[7], 59, 7)
temp2 = px_remove(temp1, 72, 34)
Gen_final_list.append(temp1)
```
