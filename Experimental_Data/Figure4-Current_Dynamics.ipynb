```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as c
import scipy.fftpack as fft
import scipy.ndimage as ndi
import scipy as sp
from PIL import Image

def px_remove(array, x, y):
    temp = array
    temp[y,x]=(temp[y+1,x]+temp[y-1,x]+temp[y,x+1]+temp[y,x-1])/4
    return temp

def px_remove_edge(array, x, y):
    temp = array
    temp[y,x]=(temp[y+1,x]+temp[y-1,x]+temp[y,x-1])/3
    return temp
```

# Load BP Toggle Data

```python
%matplotlib notebook

num = ['1','2','3','4','5','6','7','8']

BP_list = []

for i in np.arange(0,len(num),1):
    pos_im = Image.open("STXM_Data/Pulses_BP_Toggle/TIF_Data/image"+num[i]+"_pos.tif")
    neg_im = Image.open("STXM_Data/Pulses_BP_Toggle/TIF_Data/image"+num[i]+"_neg.tif")
    pos =np.array(pos_im)
    neg =np.array(neg_im)

    pos = np.rot90(pos)
    neg = np.rot90(neg)

    final = pos/neg
    final = np.fliplr(final)
    BP_list.append(final)
```
