```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as c
import scipy.fftpack as fft
import scipy.ndimage as ndi
import scipy as sp
from PIL import Image

def px_remove(array, x, y):
    temp = array
    temp[y,x]=(temp[y+1,x]+temp[y-1,x]+temp[y,x+1]+temp[y,x-1])/4
    return temp

def px_remove_edge(array, x, y):
    temp = array
    temp[y,x]=(temp[y+1,x]+temp[y-1,x]+temp[y,x-1])/3
    return temp
```

# Load Data

```python
%matplotlib notebook

fields = ['n250','n100','n80','n60','n40','n20','p0','p20','p40','p60','p80','p100','p250',]

IP_list = []

for i in np.arange(0,len(fields),1):
    pos_im = Image.open("STXM_Data/IP_Field_Sweep/TIF_Data/IP_"+fields[i]+"mT_pos.tif")
    neg_im = Image.open("STXM_Data/IP_Field_Sweep/TIF_Data/IP_"+fields[i]+"mT_neg.tif")
    pos =np.array(pos_im)
    neg =np.array(neg_im)

    pos = np.rot90(pos)
    neg = np.rot90(neg)

    final = pos/neg
    final = np.fliplr(final)
    IP_list.append(final)
```

```python
IP_final_list = []
# -250 mT #
temp1 = px_remove(IP_list[0], 170, 19)
temp2 = px_remove(temp1, 170, 20)
IP_final_list.append(temp2)

# -100 mT #
IP_final_list.append(IP_list[1])

# -80 mT #
temp1 = px_remove(IP_list[2], 26, 21)
IP_final_list.append(temp1)

# -60 mT #
IP_final_list.append(IP_list[3])

# -40 mT #
IP_final_list.append(IP_list[4])

# -20 mT #
temp1 = px_remove(IP_list[5], 100, 7)
IP_final_list.append(temp1)

# 0 mT #
IP_final_list.append(IP_list[6])

# 20 mT #
IP_final_list.append(IP_list[7])

# 40 mT #
temp1 = px_remove(IP_list[8], 139, 22)
IP_final_list.append(temp1)

# 60 mT #
IP_final_list.append(IP_list[9])

# 80 mT #
temp1 = px_remove(IP_list[10], 102, 29)
IP_final_list.append(temp1)

# 100 mT #
temp1 = px_remove(IP_list[11], 66, 8)
temp2 = px_remove(temp1, 99, 9)
IP_final_list.append(temp2)

# 250 mT #
IP_final_list.append(IP_list[12])
```
